```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv("train.csv")
data.head()
```

```python
#Filling null age values by median
data['Age'].fillna(28,inplace=True)
data.isnull().sum()
```

```python
data.shape
```

```python
dead = data['Survived'].value_counts()[0]
alive = data['Survived'].value_counts()[1]
plt.bar(['Dead','Alive'],[dead,alive],color='Green')
plt.title('Distribution of Survived and Not Survived')
plt.xlabel('Survived')
plt.ylabel('Frequency')
plt.show()
```

```python
male_cnt = data['Sex'].value_counts()['male']
female_cnt = data['Sex'].value_counts()['female']

plt.bar(['male',"female"],[male_cnt,female_cnt],color='Yellow',edgecolor='Black')
plt.title('Distribution of Genders in a Population')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
ages = data['Age']
plt.hist(ages, bins=15, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.show()
```

```python
survived_data = data[data['Survived'] == 1]
sns.countplot(x='Sex', data=survived_data)
plt.title('Number of Males and Females Who Survived')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
data['Pclass'].unique()
```

```python
total_survived = data[data['Survived']==1]
sns.countplot(x='Pclass',data=total_survived)
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Survival based on passenger ticket class')
plt.show()
```
